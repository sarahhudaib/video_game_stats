```python
import pandas as pd
import numpy as np

cols=['Rank', 'Name', 'Platform', 'Year', 'Genre', 'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
df = pd.read_csv("../input/videogamesales/vgsales.csv")[cols]
df
```

```python
df.head(10)
```

```python
# https://datascientyst.com/get-most-frequent-values-pandas-dataframe/#:~:text=To%20get%20the%20most%20frequent,It%20can%20be%20multiple%20values.
most_common_publisher=df['Publisher'].mode()
most_common_publisher[0]
```

```python
most_common_platform = df["Platform"].mode()
most_common_platform[0]
```

```python
most_common_genre = df["Genre"].mode()
most_common_genre[0]
```

```python
top_twenty_highest_grossing_games = df.head(20)
top_twenty_highest_grossing_games
```

```python
# The mid point value
na_median_sales = df["NA_Sales"].median()
na_median_sales
```

```python
# mean() The average value
ten_median_na_seller_names = df["NA_Sales"].mean()

# find standard-deviation along the given index
std_na = df["NA_Sales"].std()

std_from_mean_na = (df["NA_Sales"].head()[0] - ten_median_na_seller_names) / std_na
std_from_mean_na
```

```python
Nintendo_Wii = df[df['Platform']=='Wii']
Nintendo_Wii.describe()
Nintendo_Wii
```

```python
Nintendo_Wii = df.groupby('Publisher')
Nintendo_Wii.count()['Global_Sales']
```

```python

Nintendo_Wii=df.groupby("Platform")["Global_Sales"].mean().sort_values(ascending=False)
Nintendo_Wii
```

```python
def test():

    def assert_equal(actual,expected):
        assert actual == expected, f"Expected {expected} but got {actual}"

    assert_equal(most_common_publisher[0], 'Electronic Arts')
    assert_equal(most_common_platform[0], 'DS')
    assert_equal(most_common_genre[0], 'Action')
    assert_equal(top_twenty_highest_grossing_games.iloc[0].Name, 'Wii Sports')
    assert_equal(top_twenty_highest_grossing_games.iloc[19].Name, 'Brain Age: Train Your Brain in Minutes a Day')
    assert_equal(na_median_sales, 0.08)
    assert_equal(ten_median_na_seller_names, ten_median_na_seller_names)

    print("Success!!!")

test()
```
